```python
import pandas as pd
import numpy as np
from sklearn.preprocessing import LabelEncoder
import seaborn as sns
from matplotlib import pyplot as plt
pd.set_option('display.max_columns', 200)
pd.set_option('display.width', 1000)

clinical_P=pd.read_csv('Dataset/data_clinical_patient.txt',sep="\t")
clinical_S=pd.read_csv('Dataset/data_clinical_sample.txt',sep="\t")
```

```python
Dataset_Features = pd.merge(clinical_P, clinical_S, how='inner', on = 'PATIENT_ID')
Dataset_Features
print('Number of missing values:')
for col in Dataset_Features.columns:
    if Dataset_Features[col].isna().sum() > 19:
        print('\t%s: %d' % (col,Dataset_Features[col].isna().sum()))
```

```python
Dataset_Labels = pd.read_csv('Dataset/data_expression_median.txt',sep="\t")
Dataset_Labels
Dataset_Labels.columns=Dataset_Labels.columns.str.replace('Hugo_Symbol','PATIENT_ID')
LabelsData= Dataset_Labels.set_index('PATIENT_ID').transpose()
df = pd.DataFrame(data=LabelsData)
df.reset_index(inplace=True,col_level=0)

df.rename(columns = {"SAMPLE_ID": "NEW NAME", 
                     "index":"PATIENT_ID"}, 
                                 inplace = True) 
```

```python
df
```

```python
combined = pd.merge(df, Dataset_Features, how='inner', on ='PATIENT_ID')
combined["OS_STATUS"].head(20)
combined.describe()
print('Number of instances = %d' % (combined.shape[0]))
print('Number of attributes = %d' % (combined.shape[1]))

print('Number of missing values:')
for col in combined.columns:
    if combined[col].isna().sum() > 19:
        print('\t%s: %d' % (col,combined[col].isna().sum()))

dups = combined.duplicated()
print('Number of duplicate rows = %d' % (dups.sum()))
combined
```
